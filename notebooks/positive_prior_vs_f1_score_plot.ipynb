```python
import os
import sys
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import nci_eval.metrics as metrics
from nci_eval.plots import PositivePriorF1ScorePlot, PositivePriorPrecisionPlot
```

PositivePrevalence-Precision (P3) plots
========

```python
fig = PositivePriorPrecisionPlot() \
    .add(tpr=0.9, fpr=0.01, name="Classifier 1") \
    .add(tpr=0.6, fpr=0.001, name="Classifier 2") \
    .plot(positive_priors=np.logspace(-6, 0, num=100), figsize=(8,3))
```

```python
fig = PositivePriorPrecisionPlot() \
    .add(tpr=0.9, fpr=0.01, name="Classifier 1", sigma_tpr=0.01, sigma_fpr=0.0005) \
    .add(tpr=0.6, fpr=0.001, name="Classifier 2",  sigma_tpr=0.01, sigma_fpr=0.0005) \
    .plot(positive_priors=np.logspace(-6, 0, num=100), figsize=(8,3))
```

Max width ($\Delta$) of the UB($\eta$) - LB($\eta$) confidence interval 
====

```python
positive_prior, delta = metrics.max_width_of_precision_confidence_interval(0.6, 0.001, 0.06, 0.0005)
print("The interval is widest at positive prevalence %.5f width the width of %.3f" % (positive_prior, delta))
```
